```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个点数为 8 x 6 的窗口, 并设置分辨率为 80 像素/每英寸

plt.figure(figsize=(4, 2), dpi=80) 

# 再创建一个规格为 1 x 1 的子图

plt.subplot(111) 
x = np.linspace(-20, 10, 1000)
y1 = x**2 + 3      # 曲线 y1
y2 = 8 - x      # 曲线 y2


# 绘制颜色为蓝色、宽度为 1 像素的连续曲线 y1
plt.plot(x, y1, color="blue", linewidth=1.0, linestyle="-.") 
# 绘制颜色为紫色、宽度为 2 像素的不连续曲线 y2

plt.plot(x, y2, color="#800080", linewidth=1.0, linestyle="--") 
# 设置横轴的上下限

plt.xlim(-4, 4) 
# 设置纵轴的上下限

plt.ylim(1, 20) 
# 设置横轴精准刻度

plt.xticks(np.linspace(-4,4,9))
# 设置纵轴精准刻度
plt.yticks(np.linspace(0,20,9))
# 设置横轴标签
plt.xlabel("X")
# 设置纵轴标签
plt.ylabel("Y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.2)
y = np.sin(x) +x**2
z = np.cos(x)        
# 在第一个坐标轴上,画上正弦信号
fig, axs = plt.subplots(nrows=2, ncols=2)     

print(axs)
axs[0][0].plot(x, y)
axs[0][0].set_ylabel('Sine')

# 第二个坐标轴
axs[1][1].plot(x, z)
axs[1][1].set_ylabel('Cosine')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt 
import numpy as np 
  
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
  
# Make data 
u = np.linspace(0, 2 * np.pi, 100) 
v = np.linspace(0, np.pi, 100) 
x = 10 * np.outer(np.cos(u), np.sin(v)) 
y = 10 * np.outer(np.sin(u), np.sin(v)) 
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v)) 
  
# Plot the surface 
ax.plot_surface(x, y, z, color='red') 
  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(n):
    return np.exp(-n)*np.cos(2*np.pi*n)
x1 = np.arange(0,5,0.1)
x2=np.arange(0,5,0.01)
t1 = f(x1)
t2= f(x2)
plt.figure(1)
plt.subplot(211)
plt.title='matplotlib'
plt.plot(x1,t1,'bo',x2,t2,'k')
plt.subplot(212)

plt.plot(x2,np.cos(np.exp(x2/2)),'--')
plt.show()

```
